```python
dir = '/home/dan/serpent/frames'
file = os.listdir(dir)[0]
```

```python
im = cv2.imread(os.path.join(dir, file))
im = cv2.cvtColor(im, cv2.COLOR_RGB2BGR)
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

# (thresh, im_bw) = cv2.threshold(im, 255, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
# im = cv2.cvtColor(im_bw, cv2.COLOR_GRAY2BGR)
```

```python
image = Image.fromarray(im)
```

```python
def change_contrast(img, level):
    factor = (259 * (level + 255)) / (255 * (259 - level))
    def contrast(c):
        return 128 + factor * (c - 128)
    return img.point(contrast)
```

```python

image = change_contrast(image, -100)
```

```python
plt.imshow(image)
```
