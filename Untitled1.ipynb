```python
import cv2
import numpy as np
import os
```

```python
actions = []
observations = []
rewards = []
episode_returns = np.zeros((1,))
episode_starts = []


episode_starts.append(1)

save_dir = 'images'

dir = '/home/dan/serpent/history_processed/'
x = 0
for file in os.listdir(dir):
    if x % 1000 == 0:
        print(x)

    im = cv2.imread(os.path.join(dir, file))
    x = x + 1
    
    obs_ = cv2.cvtColor(im, cv2.COLOR_RGB2BGR)
    cv2.imwrite(os.path.join(save_dir, file), obs_)
    
    split = file.split('_')
    if split[1] == 'buy':
        action = 0
    elif split[1] == 'sell':
        action = 1
    else: action = 2
    
    reward = float(split[2].strip('.jpg'))

    actions.append(action)
    rewards.append(reward)
    episode_starts.append(0)
    
    observations.append(os.path.join('images', file))
    
observations = np.array(observations)
actions = np.array(actions).reshape((-1, 1))
rewards = np.array(rewards).reshape((-1, 1))
episode_starts = np.array(episode_starts[:-1]).reshape((-1, 1))


numpy_dict = {
        'actions': actions,
        'obs': observations,
        'rewards': rewards,
        'episode_returns': episode_returns.reshape((-1, 1)),
        'episode_starts': episode_starts
    }

for key, val in numpy_dict.items():
    print(key, val.shape)
        
np.savez('dataset', **numpy_dict)        
```

```python
from baselines.t4_env import T4Env
import os
```

```python
!export DISPLAY=:1
class PretrainEnv():
    def __init__(self):
        self._env = T4Env(fake=True)
    
    def set_files(self, files, dir):
        self.files = files
        self.x = 0
        self.dir = dir
        self.x_reward = 0.0
        self.x_action = 'hold'
        
    def step(self):
        file = self.files[self.x]
        im = cv2.imread(os.path.join(self.dir, file))
        
        ret_action = self.x_action
        ret_reward = self.x_reward
        obs = im
        
        split = file.split('_')
        self.x_reward = float(split[2].strip('.jpg'))
        
        if split[1] == 'buy':
            action = 0
        elif split[1] == 'sell':
            action = 1
        else: action = 2
        
        self.x_action = action
        
        self.x = self.x + 1
        
        return obs, ret_reward, False, {}

dir = '/home/dan/serpent/history_processed/'
files = os.listdir(dir)

env = PretrainEnv()
env.set_files(files, dir)

def model_fn(obs):
    return env.x_action
```

```python
from stable_baselines.gail import generate_expert_traj

generate_expert_traj(model_fn, 't4_dataset', env, n_episodes=1)
```
